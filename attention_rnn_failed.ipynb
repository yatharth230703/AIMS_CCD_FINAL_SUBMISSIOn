```python
import numpy as np

df['sin(Azimuth)'] = np.sin(np.radians(df['Azimuth Angle [degrees]']))
df['cos(Azimuth)'] = np.cos(np.radians(df['Azimuth Angle [degrees]']))

df.drop(columns=['Azimuth Angle [degrees]'], inplace=True)

print(df.head())
```

```python
import pandas as pd

# Assuming 'df' is your initial DataFrame

# 1. Drop rows where the target variable 'Total Cloud Cover [%]' is missing
df_cleaned = df.dropna(subset=['Total Cloud Cover [%]'])

# 2. Forward fill missing values
df_ffill = df_cleaned.fillna(method='ffill')

# 3. Backward fill any remaining missing values
df_imputed = df_ffill.fillna(method='bfill')

# Final check for remaining missing values
remaining_missing_values = df_imputed.isna().sum()
print("Remaining missing values after forward and backward fill:")
print(remaining_missing_values)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense



# Normalize the data
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_imputed)

def create_sequences(data, sequence_length):
    X = []
    y = []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i+sequence_length])
        y.append(data[i+sequence_length, -1])
    return np.array(X), np.array(y)

sequence_length = 5
X, y = create_sequences(df_scaled, sequence_length)

split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Check the shapes
X_train.shape, X_test.shape, y_train.shape, y_test.shape

```

```python
class AttentionLayer(tf.keras.layers.Layer):
    def __init__(self, **kwargs):
        super(AttentionLayer, self).__init__(**kwargs)

    def build(self, input_shape):
        self.W = self.add_weight(name='attention_weight', shape=(input_shape[-1], input_shape[-1]), initializer='random_normal', trainable=True)
        self.b = self.add_weight(name='attention_bias', shape=(input_shape[-1],), initializer='zeros', trainable=True)
        super(AttentionLayer, self).build(input_shape)

    def call(self, inputs):
        score = tf.nn.tanh(tf.tensordot(inputs, self.W, axes=[-1, 0]) + self.b)
        attention_weights = tf.nn.softmax(score, axis=1)
        context_vector = attention_weights * inputs
        context_vector = tf.reduce_sum(context_vector, axis=1)
        return context_vector
```

```python
input_seq = Input(shape=(sequence_length, X_train.shape[2]))
lstm_out = LSTM(64, return_sequences=True)(input_seq)
attention_out = AttentionLayer()(lstm_out)

output = Dense(1)(attention_out)
model = Model(inputs=input_seq, outputs=output)
model.compile(optimizer='adam', loss='mse')
model.summary()
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(np.concatenate((np.zeros((predictions.shape[0], df.shape[1] - 1)), predictions), axis=1))[:, -1]

#predictions for 15, 25, and 30 minutes intervals
forecast_horizons = [15, 25, 30]
forecasted_values = predictions[forecast_horizons]

print(f"Predictions for {forecast_horizons} minute intervals: {forecasted_values}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
y_test_actual = scaler.inverse_transform(np.concatenate((np.zeros((y_test.shape[0], df.shape[1] - 1)), y_test.reshape(-1, 1)), axis=1))[:, -1]
mse = mean_squared_error(y_test_actual, predictions)
r2 = r2_score(y_test_actual, predictions)
mae = mean_absolute_error(y_test_actual, predictions)
print(f"MSE: {mse}")
print(f"R2: {r2}")
print(f"MAE: {mae}")
```

```python
# unidirectional RNN model
input_seq_rnn = Input(shape=(sequence_length, X_train.shape[2]))
rnn_out = tf.keras.layers.SimpleRNN(64, return_sequences=False)(input_seq_rnn)
output_rnn = Dense(1)(rnn_out)
model_rnn = Model(inputs=input_seq_rnn, outputs=output_rnn)

model_rnn.compile(optimizer='adam', loss='mse')
model_rnn.summary()


model_rnn.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

predictions_rnn = model_rnn.predict(X_test)

predictions_rnn = scaler.inverse_transform(np.concatenate((np.zeros((predictions_rnn.shape[0], df.shape[1] - 1)), predictions_rnn), axis=1))[:, -1]
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
y_test_actual = scaler.inverse_transform(np.concatenate((np.zeros((y_test.shape[0], df.shape[1] - 1)), y_test.reshape(-1, 1)), axis=1))[:, -1]
mse = mean_squared_error(y_test_actual, predictions_rnn)
r2 = r2_score(y_test_actual, predictions_rnn)
mae = mean_absolute_error(y_test_actual, predictions_rnn)
print(f"MSE: {mse}")
print(f"R2: {r2}")
print(f"MAE: {mae}")
```

```python
# Average the predictions from both models
predictions_ensemble = (predictions + predictions_rnn) / 2

# Calculate evaluation metrics for the ensemble
mse_ensemble = mean_squared_error(y_test_actual, predictions_ensemble)
r2_ensemble = r2_score(y_test_actual, predictions_ensemble)
mae_ensemble = mean_absolute_error(y_test_actual, predictions_ensemble)

print(f"Ensemble MSE: {mse_ensemble}")
print(f"Ensemble R2: {r2_ensemble}")
print(f"Ensemble MAE: {mae_ensemble}")

# Get predictions for 15, 25, and 30 minutes intervals for the ensemble
forecasted_values_ensemble = predictions_ensemble[forecast_horizons]

print(f"Ensemble Predictions for {forecast_horizons} minute intervals: {forecasted_values_ensemble}")
```
