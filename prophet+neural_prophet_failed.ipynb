```python
import pandas as pd
import numpy as np
from prophet import Prophet

# Load your preprocessed data
df = df_imputed  # Adjust the path to your data file

# Ensure the date column is of datetime type
#df['DATE (MM/DD)'] = pd.to_datetime(df['DATE (MM/DD)'], format='%m/%d')

# Prepare the dataframe for Prophet
df = df.rename(columns={'DATE (MM/DD)': 'ds', 'Total Cloud Cover [%]': 'y'})

# Create sequences of length 5
def create_sequences(data, seq_length=5):
    sequences = []
    targets = []
    for i in range(len(data) - seq_length):
        seq = data.iloc[i:i + seq_length].copy()
        target = data['y'].iloc[i + seq_length]
        sequences.append(seq)
        targets.append(target)
    return sequences, targets

sequences, targets = create_sequences(df)

# Initialize a list to store forecasts
forecasts = []

# Fit the Prophet model on each sequence
for seq in sequences:
    seq_df = pd.DataFrame(seq)
    model = Prophet()
    model.fit(seq_df)
    future = model.make_future_dataframe(periods=1)
    forecast = model.predict(future)
    forecasts.append(forecast['yhat'].iloc[-1])

# Evaluate the model performance
targets = np.array(targets)
forecasts = np.array(forecasts)

mae = np.mean(np.abs(targets - forecasts))
mse = np.mean((targets - forecasts) ** 2)
rmse = np.sqrt(mse)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")

# Plot the results
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(targets, color='blue', label='Actual')
plt.plot(forecasts, color='red', label='Predicted')
plt.title('Total Cloud Cover Prediction')
plt.xlabel('Time')
plt.ylabel('Total Cloud Cover [%]')
plt.legend()
plt.show()
```

```python
import pandas as pd
from neuralprophet import NeuralProphet
import matplotlib.pyplot as plt

# Load your preprocessed data
df = df_imputed  # Adjust the path to your data file

# Ensure the date column is of datetime type
#df['DATE (MM/DD)'] = pd.to_datetime(df['DATE (MM/DD)'], format='%m/%d')

# Prepare the dataframe for NeuralProphet
df = df.rename(columns={'DATE (MM/DD)': 'ds', 'Total Cloud Cover [%]': 'y'})

#nlags=5 matlab sequence  5
model = NeuralProphet(n_lags=5, n_forecasts=1)

# Fit the model to your data
model.fit(df, freq='D')

# Make a dataframe for future dates
future = model.make_future_dataframe(df, periods=30)

# Make predictions
forecast = model.predict(future)

# Plot the forecast
fig_forecast = model.plot(forecast)
plt.show()

# Plot forecast components
fig_components = model.plot_components(forecast)
plt.show()

# Plot model parameters
fig_parameters = model.plot_parameters()
plt.show()

```
