```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
threshold = 0.90 * len(data)
data_reduced = data.dropna(thresh=threshold, axis=1)

imputer = SimpleImputer(strategy='median')
data_imputed = pd.DataFrame(imputer.fit_transform(data_reduced), columns=data_reduced.columns)

scaler = StandardScaler()
features_scaled = scaler.fit_transform(data_imputed.drop(columns=['Total Cloud Cover [%]']))

data_processed = pd.DataFrame(features_scaled, columns=data_imputed.columns.drop('Total Cloud Cover [%]'))
data_processed['Total Cloud Cover [%]'] = data_imputed['Total Cloud Cover [%]']

data_processed.head(100)
```

```python
data_processed.columns = [col.replace('[', '_').replace(']', '_') for col in data_processed.columns]
data_processed.head()
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
X = data_processed.drop(columns=['Total Cloud Cover _%_'])

train_size = int(0.8 * len(X))

X_train, X_test = X[:train_size], X[train_size:]

model = IsolationForest(random_state=42)
model.fit(X_train)

y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
n_errors_train = (y_pred_train == -1).sum()
n_errors_test = (y_pred_test == -1).sum()
print(f'Number of anomalies in the training set: {n_errors_train}')
print(f'Number of anomalies in the test set: {n_errors_test}')
y_test_true = np.ones_like(y_pred_test)
accuracy = accuracy_score(y_test_true, y_pred_test)
mse = mean_squared_error(y_test_true, y_pred_test)
mae = mean_absolute_error(y_test_true, y_pred_test)
r2 = r2_score(y_test_true, y_pred_test)

print(f'accuracy: {accuracy}')
print(f'mse: {mse}')
print(f'mae: {mae}')
print(f'r2_score: {r2}')
```
